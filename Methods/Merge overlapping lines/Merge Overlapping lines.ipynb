The third image overlays both the roads one on top of another. We want to merge these overlaid layers into a single layer. Why?

1. Overlaid layers will give faulty statistics when we calcualte. Consider a small area in which we want to calculate the length of the road. We don't want to calculate the same road twice.

2. Size of the file increases due to this duplication.

# Just Dissolve won't help

```python
gdf.dissolve().plot()
```

`Dissolve` does help in avoiding the duplication error. So the statistical errors won't happen. But, the size of the file will not reduce with it. This is because it almost joins the vertices of both lines together to form a single line.

Eg: If line A has 50 points; line B has 50 different points -- Dissolved line will have 100 points. (If there are common points between line A and B, then this number might reduce -- only one of the points will be in the dissolved line)

```python
gdf.count_coordinates()
```

```python
gdf.count_coordinates().sum()
```

```python
gdf.dissolve().count_coordinates()
```

Since the total number of points didn't reduce much, file size will also not reduce.

Hence use `line_merge`. It is an interesting function that tries to convert contiguos lines into a single line. 

```python
from shapely.geometry import MultiLineString, Point
s = gpd.GeoSeries(
    [
        MultiLineString([[(0, 2), (0, 10)], [(0, 10), (5, 10)]]),
        MultiLineString([[(0, 2), (0, 10)], [(0, 11), (5, 10)]]),
        MultiLineString(),
        MultiLineString([[(0, 0), (1, 0)], [(0, 0), (3, 0)]]),
        Point(0, 0),
    ])
s.line_merge()
```

```python
s.line_merge()[3]
```

```python
s[3]
```

Can you see that (0,0) to (0,1) part is thicker originally and not after using line_merge! So it merges the contiguous lines and removed the overlap.

```python
gdf.dissolve().line_merge().plot()
```

```python
gdf.dissolve().line_merge().count_coordinates()
```

Less points! So the size of the file will be lesser!

If you further want to reduce the size of the image, you can use simplify.

```python
gdf.dissolve().line_merge().simplify(0.01).plot()
```

```python
gdf.dissolve().line_merge().simplify(0.01).count_coordinates()
```
